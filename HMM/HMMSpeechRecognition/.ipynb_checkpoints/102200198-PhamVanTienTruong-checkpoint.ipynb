## Bài 5.2

Độ chính xác lớn nhất trước khi chuẩn hóa là *0.6735074626865672* với Phương pháp làm sạch Mean và có xử lí ngoại lệ

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import PowerTransformer

import scipy.stats as stat
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np;

X_train = pd.DataFrame()
X_test = pd.DataFrame()

NameMin = ''
NameMax = ''
Max = -np.inf
Min = np.inf

Gt = []
```

Danh sách các đặc trưng bị trống

```python
df1 = pd.read_csv('titanic.csv', usecols=['Age', 'Fare', 'Survived' , 'Sex' , 'SibSp' , 'Parch','Pclass'])
df1['newSex'] = [1 if i == 'male' else 0 for i in df1.Sex]
df1.isnull().sum()
```

Các Hàm làm sạch dữ liệu

```python
def mean_NA_repace():
    mean = X_train.Age.mean()
    X_train["Age"].fillna(mean, inplace=True)
    X_test["Age"].fillna(mean, inplace=True)

def outlier():

    uppper_boundary = X_train['Age'].mean() + 3 * X_train['Age'].std()
    lower_boundary = X_train['Age'].mean() - 3 * X_train['Age'].std()

    X_train.loc[X_train['Age'] >= uppper_boundary, 'Age'] = uppper_boundary
    X_test.loc[X_test['Age'] >= uppper_boundary, 'Age'] = uppper_boundary


    IQR = X_train.Fare.quantile(0.75) - X_train.Fare.quantile(0.25)
    lower_bridge = X_train['Fare'].quantile(0.25) - (IQR * 3)
    upper_bridge = X_train['Fare'].quantile(0.75) + (IQR * 3)

    X_train.loc[X_train['Fare'] >= upper_bridge, 'Fare'] = upper_bridge
    X_test.loc[X_test['Fare'] >= upper_bridge, 'Fare'] = upper_bridge

def NoneFunc():
    pass
```

Các Hàm chuẩn hóa dữ liệu

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist() # histogram
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=plt)# prob plot
    plt.show()


def standardscaler():
    global X_train
    global X_test

    scaler = StandardScaler().set_output(transform="pandas").fit(X_train)

    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

def minmaxscaler():
    global X_train
    global X_test

    min_max = MinMaxScaler().set_output(transform="pandas").fit(X_train)
    X_train = min_max.transform(X_train)
    X_test = min_max.transform(X_test)


def robustscaler():
    global X_train
    global X_test

    scaler = RobustScaler().set_output(transform="pandas").fit(X_train)

    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

def  maxabsscaler():
    global X_train
    global X_test

    scaler = MaxAbsScaler().set_output(transform="pandas").fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)


def transformationlog1p():

    X_train.Fare = np.log1p(X_train['Fare'])
    X_test.Fare = np.log1p(X_test['Fare'])

def transformationboxcox():
    X_train['Fare'], parameters = stat.boxcox(X_train['Fare'] + 1)
#     # plot_data(X_train, "Fare")
    X_test['Fare'], parameters = stat.boxcox(X_test['Fare'] +1)



```

Tiến hành làm sạch dữ liệu và chuẩn hóa cho hai đặc trưng 'Age' và 'Fare'

```python
X = [standardscaler,minmaxscaler,maxabsscaler,robustscaler]
Y = [transformationboxcox,transformationlog1p]
Max = -np.inf
NameMax = ""
for i in X:
    for j in Y:
        sum = 0
        for z in range(10):
            New_X_train, New_X_test, y_train, y_test = train_test_split(df1[['Age', 'Fare']], df1['Survived'],
                                                                        test_size=0.3, random_state=z)
            X_train = pd.DataFrame(New_X_train)
            X_test = pd.DataFrame(New_X_test)

            mean_NA_repace()
            outlier()

            j()
            i()
            classifier = LogisticRegression()
            classifier.fit(X_train, y_train)
            y_pred = classifier.predict(X_test)
            sum = sum + accuracy_score(y_test, y_pred)

        Gt.append(sum/10)
        funcname = [i.__str__().replace("<function ","").split(" at ")[0],j.__str__().replace("<function ","").split(" at ")[0]]
        print( "| {:<20} | {:<25} | {:<20} |".format(funcname[0],funcname[1],sum/10))

        if sum/10 > Max:
            Max = sum/10
            NameMax = funcname

```

Cánh chuẩn hóa cho giá trị lớn nhất là :

```python
print(NameMax , " :" ,Max)
```

Từ kết quả cho thấy sau khi chuẩn hóa thì độ chính được cải thiện rất ít , chỉ khoảng 1%

Sử dụng thêm kĩ thuật lựa chọn đặc trưng để cải thiện độ chính xác :

```python
sum = 0
Max = -np.inf
NameMax = ""
for i in X:
    for j in Y:
        sum = 0
        for z in range(10):
            New_X_train, New_X_test, y_train, y_test = train_test_split(df1[['Age', 'Fare','newSex' ,'Parch','SibSp','Pclass']], df1['Survived'],
                                                                test_size=0.3, random_state=z)
            X_train = pd.DataFrame(New_X_train)
            X_test = pd.DataFrame(New_X_test)


            mean_NA_repace()
            outlier()

            j()
            i()


            X_train['Parch'] = X_train['Parch'].transform(lambda x: np.log1p(x))
            X_train['SibSp'] = X_train['SibSp'].transform(lambda x: np.log1p(x))

            X_test['Parch'] = X_test['Parch'].transform(lambda x: np.log1p(x))
            X_test['SibSp'] = X_test['SibSp'].transform(lambda x: np.log1p(x))


            from sklearn.feature_selection import SequentialFeatureSelector
            from sklearn.linear_model import LogisticRegression


            logit = LogisticRegression()
            selector = SequentialFeatureSelector(logit, n_features_to_select=4, )

            selector.fit(X_train, y_train)

            X_train = selector.set_output(transform="pandas").transform(X_train)
            X_test = selector.set_output(transform="pandas").transform(X_test)

            classifier = LogisticRegression()
            classifier.fit(X_train, y_train)
            y_pred = classifier.predict(X_test)

            sum = sum + accuracy_score(y_test, y_pred)
        funcname = [i.__str__().replace("<function ","").split(" at ")[0],j.__str__().replace("<function ","").split(" at ")[0]]
        print( "| {:<20} | {:<25} | {:<20} |".format(funcname[0],funcname[1],sum/10))

        if sum/10 > Max:
            Max = sum/10
            NameMax = funcname

print()
print(NameMax , " :" ,Max)
```

## Bài 8.1

```python
Max = -np.inf
NameMax = ""

scaler = [StandardScaler().set_output(transform="pandas"),MinMaxScaler().set_output(transform="pandas"),RobustScaler().set_output(transform="pandas"), MaxAbsScaler().set_output(transform="pandas")]
transformer = [transformationboxcox,transformationlog1p]
scalername = ['StandardScaler' ,'MinMaxScaler','RobustScaler','MaxAbsScaler']
from sklearn.model_selection  import cross_validate
from sklearn.pipeline import make_pipeline


X_train = pd.DataFrame(df1[['Age', 'Fare','newSex' ,'Parch','SibSp','Pclass']])
y_train = df1['Survived']

X_test = pd.DataFrame(df1[['Age', 'Fare','newSex' ,'Parch','SibSp','Pclass']])
y_test = df1['Survived']

for idx,i in enumerate(scaler):
    for idy,j in enumerate(transformer):
        mean_NA_repace()
        outlier()

        j()
        X_train['Parch'] = X_train['Parch'].transform(lambda x: np.log1p(x))
        X_train['SibSp'] = X_train['SibSp'].transform(lambda x: np.log1p(x))


        X = X_train
        y = y_train

        logit = LogisticRegression()
        selector = SequentialFeatureSelector(logit, n_features_to_select=4, )
        pipe = make_pipeline(i, selector, logit)

        result = cross_validate(pipe, X, y ,cv=5)

        funcname = j.__str__().replace("<function ","").split(" at ")[0]
        print( "| {:<20} | {:<25} | {:<30} | {:<20} ".format(scalername[idx],funcname,result['test_score'].__str__(),result['test_score'].sum()/5))

        if result['test_score'].sum()/5 > Max:
            Max = result['test_score'].sum()/5
            NameMax = scalername[idx] + '+' +funcname

print()
print(NameMax , " :" ,Max)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection  import cross_validate

New_X_train, New_X_test, y_train, y_test = train_test_split(df1[['Age', 'Fare','newSex' ,'Parch','SibSp','Pclass']], df1['Survived'],
                                                                test_size=0.3, random_state=10, stratify=df1['Survived'])
X_train = pd.DataFrame(New_X_train)
X_test = pd.DataFrame(New_X_test)

train_classes = pd.Series(y_train).value_counts(normalize=True)
test_classes = pd.Series(y_test).value_counts(normalize=True)

print('Phân bố các lớp trong tập huấn luyện:\n', train_classes)
print('Phân bố các lớp trong tập kiểm tra:\n', test_classes)

mean_NA_repace()
outlier()
transformationboxcox()
standardscaler()


penalty = ['l1', 'l2', 'none']
C = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
hyperparameters = dict(penalty=penalty, C=C)

model = LogisticRegression()
grid_search = GridSearchCV(model, hyperparameters, cv=5)
grid_search.fit(X_train, y_train)

print('Siêu tham số tốt nhất:', grid_search.best_params_)
print('Điểm số tốt nhất:', grid_search.best_score_)

classifier =  LogisticRegression(penalty='l2',C=0.01)

logit = LogisticRegression()
selector = SequentialFeatureSelector(logit, n_features_to_select=4,cv=5 )
selector.fit(X_train,y_train)

X_train = selector.set_output(transform="pandas").transform(X_train)
X_test = selector.set_output(transform="pandas").transform(X_test)


classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print('accuracy_score:',accuracy_score(y_test, y_pred))
```
